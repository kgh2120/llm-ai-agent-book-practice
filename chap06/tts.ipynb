```python
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key = api_key)

```

```python
response = client.audio.speech.create(
    model="tts-1-hd",
    voice="alloy",
    input="Hello word! This is a TTS test."
)
response.write_to_file("hello_world.mp3")

import IPython.display as ipd

ipd.Audio("hello_world.mp3")
```

```python
voice = "ash"
mp3_file = f"hello_word_{voice}.mp3"

response = client.audio.speech.create(
    model="tts-1-hd",
    voice=voice,
    input=f"Hello word! I'm {voice}. This is a TTS test."
)

response.write_to_file(mp3_file)

import IPython.display as ipd

ipd.Audio(mp3_file)
```

```python
import json

with open('./image/image_quiz_eng.json', 'r', encoding='utf-8') as f:
    eng_dict = json.load(f)

eng_dict
```

```python
voices = ['alloy', 'ash', 'coral', 'echo', 'fable', 'onyx', 'nova', 'sage']

for q in eng_dict:
    no = q['no']
    quiz = q['eng']
    quiz = quiz.replace("- (1)", "- One.\t")
    quiz = quiz.replace("- (2)", "- Two.\t")
    quiz = quiz.replace("- (3)", "- Three.\t")
    quiz = quiz.replace("- (4)", "- Four.\t")

    print(no, quiz)

    voice = voices[no % len(voices)]

    response = client.audio.speech.create(
        model="tts-1-hd",
        voice=voice,
        input=f'#{no}. {quiz}'
    )

    response.write_to_file(f"./audio/{no}.mp3")
```
